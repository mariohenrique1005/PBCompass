### Selecionando os 5 aplicativos mais instalados e plotando no gráfico

```python
df4=df4.dropna(subset=['Installs'])
top_apps=df4.nlargest(5,'Installs')
plt.figure(figsize=(10,5))
plt.bar(top_apps['App'], top_apps['Installs'], color='green')
plt.title('5 Apps mais instalados')
plt.xlabel('App')
plt.ylabel('Número de Instalações (bilhões)')
plt.xticks(rotation=90)
plt.show()
```

### Contando a frequencia de cada categoria e depois plotando em um gráfico de pizza de acordo com sua frequência

```python
df4=df4.dropna(subset=['Category'])
category_contagem = df4['Category'].value_counts()
plt.figure(figsize=(14, 14))
plt.pie(category_contagem, labels=category_contagem.index, autopct='%1.1f%%', startangle=180, colors=plt.cm.Paired.colors)
plt.title('Distribuição de Categorias de Apps')
plt.show()
```

### Criando novo gráfico para top 10 apps relativos a reviews e rating

```python
df_unique = df4[['App', 'Reviews', 'Rating']].drop_duplicates(subset='App')
top_10_unique = df_unique.sort_values(by=['Reviews', 'Rating'], ascending=False).head(10)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(top_10_unique['Reviews'], top_10_unique['Rating'], color='blue', s=100, alpha=0.7)
plt.title('Top 10 Aplicativos por Número de Reviews e Rating', fontsize=14)
plt.xlabel('Número de Reviews', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
